```python
pd.unique(car_df.fueltype) #'gas', 'diesel'
car_df.loc[car_df.fueltype=='diesel','fueltype'] = 0
car_df.loc[car_df.fueltype=='gas','fueltype'] = 1
pd.unique(car_df.fueltype)
```

```python
pd.unique(car_df.aspiration) #'std', 'turbo'
car_df.loc[car_df.aspiration=='std','aspiration'] = 0
car_df.loc[car_df.aspiration=='turbo','aspiration'] = 1
pd.unique(car_df.aspiration)
```

```python
pd.unique(car_df.enginelocation) #'front', 'rear'
car_df.loc[car_df.enginelocation=='front','enginelocation'] = 0
car_df.loc[car_df.enginelocation=='rear','enginelocation'] = 1
pd.unique(car_df.enginelocation)
```

```python
pd.unique(car_df.drivewheel)
car_df.loc[car_df.drivewheel=='fwd','drivewheel'] = 0
car_df.loc[car_df.drivewheel=='rwd','drivewheel'] = 1
pd.unique(car_df.drivewheel)
```

```python
#Dummy variable list : CarName,carbody,fuelsystem,enginetype
dummy = pd.get_dummies(car_df[['CarName','carbody','fuelsystem','enginetype']],drop_first=True)
dummy.head()
```

```python
car_final = pd.concat([car_df,dummy], axis=1)
car_final.head()
```

```python
#Time to remove the categorical columns
car_final.drop(['CarName','carbody','fuelsystem','enginetype'],axis=1,inplace=True)
car_final.head()
```

```python
#Splitting data into train and test
from sklearn.model_selection import train_test_split
car_train,car_test = train_test_split(car_final,train_size=0.7, test_size=0.3, random_state=100)
car_test.count()
```

```python
#Time to scale the train data minus the 0,1 columns
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_vars = ['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price','doornumber','cylindernumber']

car_train[num_vars] = scaler.fit_transform(car_train[num_vars])
car_train.head()
```

```python
y_train = car_train.pop('price')
x_train = car_train
```

```python
x_train.head()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
rfe = RFE(lm,10)
rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns, rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
x_train_rfe = x_train[col]
x_train_rfe.head()
```

```python
import statsmodels.api as sm
sm.add_constant(x_train_rfe)
sm
```

```python
lm = sm.OLS(y_train, x_train_rfe).fit()
print(lm.summary())
#enginetype_l
```

```python
x_train_rfe.pop('enginetype_l')
lm = sm.OLS(y_train, x_train_rfe).fit()
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = x_train_rfe.columns
vif
```

```python
vif['VIF'] = [variance_inflation_factor(x_train_rfe.values,i) for i in range(x_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
x_train_rfe.pop('carlength')
lm = sm.OLS(y_train, x_train_rfe).fit()
print(lm.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = x_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(x_train_rfe.values,i) for i in range(x_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
x_train_rfe.pop('carwidth')
lm = sm.OLS(y_train, x_train_rfe).fit()
print(lm.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = x_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(x_train_rfe.values,i) for i in range(x_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
#Looks great
```

```python
#Residual analysis
y_train_pred = lm.predict(x_train_rfe)
y_train_pred.head()
```

```python
sns.distplot((y_train_pred-y_train),bins=30)
plt.xlabel('Error Terms')
#Normalized so no hidden issues
```
